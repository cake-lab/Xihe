```python
from mpl_toolkits import mplot3d
```

```python
%matplotlib inline
import math
import numpy as np
import pyreality as pr
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(5, 5))

ax = plt.axes(projection='3d')
ax.set_box_aspect([1, 1, 1])

# Create a sphere
r = 1
pi = np.pi
cos = np.cos
sin = np.sin
phi, theta = np.mgrid[0.0:pi:100j, 0.0:2.0*pi:100j]
x = r*sin(phi)*cos(theta)
y = r*sin(phi)*sin(theta)
z = r*cos(phi)

ax.plot_surface(
    x, y, z,  rstride=1, cstride=1, color='lightskyblue', alpha=0.3, linewidth=0)

pos = pr.fibonacci_sphere(128)
uniform_points = pos
xx = pos[:, 0]
yy = pos[:, 1]
zz = pos[:, 2]

ax.scatter(xx,yy,zz,color="k",s=20)

ax.grid(False) 
# ax.set_xlim([-1,1])
# ax.set_ylim([-1,1])
# ax.set_zlim([-1,1])

# ax.set_xticks([-1, -0.5, 0, 0.5, 1])
# ax.set_yticks([-1, -0.5, 0, 0.5, 1])
# ax.set_zticks([-1, -0.5, 0, 0.5, 1])

ax.xaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
ax.yaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
ax.zaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))

# ax.xaxis.pane.set_edgecolor('lavender')
# ax.xaxis.pane.set_facecolor('lavender')
# ax.yaxis.pane.set_edgecolor('lavender')
# ax.yaxis.pane.set_facecolor('lavender')
# ax.zaxis.pane.set_edgecolor('lavender')
# ax.zaxis.pane.set_facecolor('lavender')
ax.set_axis_off()

plt.tight_layout(rect=(0, 0.05, 1, 1))
plt.savefig(f'{WORKING_DIR}/figures/uniform_points.pdf')
```
